```python
df=df.drop(["Id"],axis=1)
df.head()
```

```python
x = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix as cm
```

```python
import  sklearn.metrics
```

```python
lbs = ['Iris-versicolor','Iris-setosa','Iris-virginica']
cm=sklearn.metrics.confusion_matrix(y_test, y_pred, labels=lbs)
print("Confusion metrics for GaussianNB : \n",cm )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True, xticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica'], yticklabels=['Iris-setosa','Iris-versicolor','Iris-virginica'])
plt.show()
```

```python
TP = cm[0][0]
TN = cm[1][1] + cm[1][2] + cm[2][2]
FP = cm[1][0] + cm[2][0]
FN = cm[0][1] + cm[0][2]

total = TP + TN + FN + FP

print("Confusion Matrix:")
print(cm)
print("\nTrue Positive(TP):", TP)
print("\nFalse Positive(TP):", FP)
print("\nTrue Negative(TP):", TN)
print("\nFalse Negative(TP):", FN)

```

```python
Accuracy = (TP + TN)/(total)
precision = (TP)/(TP+FP)
recall = (TP)/(TP+FN)
f1_score = 2*(precision * recall)/(precision + recall)
error = (FP+FN)/(total)

print("Acurracy of the model= ", Accuracy*100, '%')
print("Precision= ", precision)
print("Recall(sensitivity , hit rate)= ", recall)
print("f1_score= ", f1_score*100, '%')
print("Error Rate= ", error)
```

```python
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
lbs=['Iris-versicolor', 'Iris-setosa', 'Iris-virginica']
cm1 = sklearn.metrics.confusion_matrix(y_test, y_pred, labels=lbs)
print(sklearn.metrics.confusion_matrix(y_test, y_pred, labels=lbs))
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(cm1, annot=True, xticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], yticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```

```python
TP = cm1[0][0]
TN = cm1[1][1] + cm1[1][2] + cm1[2][1] + cm1[2][2]
FP = cm1[1][0] + cm1[2][0]
FN = cm1[0][1] + cm1[0][2]

total = TP+TN+FP+FN

print("True Positive (TP)",TP)
print("True Negative (TP)",TN)
print("False Positive (TP)",FP)
print("False Negative (TP)",FN)
```

```python
Accuracy  = (TP) / (TP+FP)
Recall    = (TP) / (TP+FN)
Precision = (TP) / (TP+FP)
Error     = (FP+FN) / (total)
F1_score  = 2 * (Precision * Recall) / (Precision + Recall)

print("Accuracy:",Accuracy*100,'%')
print("Precision:",Precision*100,'%')
print("Recall(Sensitvity,hit rate):",Recall*100,'%')
print("F1_Score:",F1_score*100,'%')
print("Error_rate:",Error*100,'%')

```
